```python
from pulp import *
```

```python
#Definir el problema y si es de max o min
problema = LpProblem("Prolineal", LpMaximize)

#Definir variables
x1 = LpVariable("x1", lowBound = 0)
x2 = LpVariable("x2", lowBound = 0)

#Función objetivo
problema += -15*x1-20*x2

#Restricciones
problema += -6*x1-3*x2 <= -3
problema += -9*x1-3*x2 <= -3
problema += -25*x1-25*x2 <= -21

#Resolver problema
problema.solve()

print(f"El valor optimo se obtiene cuando x1 = {x1.varValue} y x2 = {x2.varValue}, y el valor optimo es: {value(problema.objective)}")
```

```python
#Definir el problema y si es de max o min
problema2 = LpProblem("Prolineal", LpMinimize)

#Definir variables
x1_1 = LpVariable("x1_1", lowBound = 0)
x1_2 = LpVariable("x1_2", lowBound = 0)
x1_3 = LpVariable("x1_3", lowBound = 0)
x2_1 = LpVariable("x2_1", lowBound = 0)
x2_2 = LpVariable("x2_2", lowBound = 0)
x2_3 = LpVariable("x2_3", lowBound = 0)

#Función objetivo
problema2 += 5*x1_1+7*x1_2+3*x1_3+4*x2_1+6*x2_2+2*x2_3

#Restricciones
problema2 += x1_1+x2_1 <= 50
problema2 += x1_2+x2_2 <= 70
problema2 += x1_3+x2_3 <= 90
problema2 += x1_1+x1_2+x1_3 >= 120
problema2 += x2_1+x2_2+x2_3 >= 80

#Resolver problema
problema2.solve()

print(f"El valor optimo se obtiene cuando x1_1 = {x1_1.varValue} y x1_2 = {x1_2.varValue} y x1_3 = {x1_3.varValue} x2_1 = {x2_1.varValue} y x2_2 = {x2_2.varValue} y x2_3 = {x2_3.varValue}, y el valor optimo es: {value(problema2.objective)}")
```
